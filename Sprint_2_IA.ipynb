# Importação das dependências

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import classification_report

import random
```

# Analise Explanatória

```python
data = pd.read_csv('data.csv', sep=",")
```

```python
data
```

## Manipulação dos Dados

```python
scaler = StandardScaler()
```

```python
data['post-views'] = scaler.fit_transform(data[['post-views']])
data['post-reposts'] = scaler.fit_transform(data[['post-reposts']])
data['post-comments'] = scaler.fit_transform(data[['post-comments']])
data['post-likes'] = scaler.fit_transform(data[['post-likes']])
```

```python
encoder = LabelEncoder()
```

```python
data['origin'] = encoder.fit_transform(data['origin'])
data['score'] = encoder.fit_transform(data['score'])
```

```python
data.head()
```

```python
x = data.drop(['score', 'hashtag'], axis=1)
y = data['score']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
```

```python
y_pred = lda.predict(x_test)
```

```python
y_pred
```

```python
print(classification_report(y_test, y_pred))
```
